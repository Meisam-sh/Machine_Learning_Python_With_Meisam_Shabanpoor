```python
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x = np.column_stack([train.Production_Index, train.Mean_of_Hatching, train.chicken_weight_Mean, train.chickenCost_Per_kg, train.Reduction, train.Cost_b4_slaughtering_kg, train.TotalCost_b4_slaughtering, train.Income])
train_y = np.asanyarray(train.Profit_margin.values.reshape(-1, 1))
###########################################
test_x = np.column_stack([test.Production_Index, test.Mean_of_Hatching, test.chicken_weight_Mean, test.chickenCost_Per_kg, test.Reduction, test.Cost_b4_slaughtering_kg, test.TotalCost_b4_slaughtering, test.Income])#.reshape(-1, 1)  
test_y = np.asanyarray(test.Profit_margin.values.reshape(-1, 1))
```

```python

# Assuming your 'clf2' object has the necessary coefficients  
coef = clf2.coef_  

# Create a range of values for the x-axis  
xx = np.arange(np.min(train.Production_Index), np.max(train.Production_Index), 2)  

# Calculate the predicted y-values using the polynomial equation  
yy = clf2.intercept_  
for i, c in enumerate(coef):  
    yy += c * (xx ** (i+1))  

# Plot the regression line  
plt.figure(figsize=(11, 6))  
scatter = plt.scatter(train['Production_Index'], train['Profit_margin'], c=train['Reduction'], cmap='viridis', alpha=0.6)  
plt.plot(xx, yy, '-r', label='Regression Line')  
plt.xlabel('Production Index')  
plt.ylabel('Profit Margin')  
plt.title('Polynomial Regression (Degree {})'.format(len(coef)))  
plt.legend()  
plt.ylim(np.min(train['Profit_margin']),np.max(train['Profit_margin']))  # Set the y-axis limits   
plt.xlim(200,375)
plt.show()
```

```python
from sklearn.metrics import r2_score  

# تبدیل داده های تست به داده های چندجمله ای  
test_x_poly = poly.fit_transform(test_x)  
# پیش بینی مقادیر y  
test_y_ = clf.predict(test_x_poly)  

scores = cross_val_score(clf, test_y, test_y_, cv=5, scoring='r2')  

r2_mean = np.mean(scores)  
r2_std = np.std(scores)  

print(f'Mean R2-squared: {r2_mean:.2f} +/- {r2_std:.2f}')
print("_________________")

# محاسبه و چاپ خطاها  
print(f'Mean Absolute Error: {np.mean(np.abs(test_y_ - test_y))}')  
print(f'MSE: {np.mean((test_y_ - test_y) ** 2)}')  
print(f'The r2_score is: {r2_score(test_y, test_y_)}')
```
